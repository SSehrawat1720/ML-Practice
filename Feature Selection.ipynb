```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Users\SS\Downloads\MDS Course files\Data set\drug200.csv")
df.head()
```

```python
from sklearn import preprocessing
df=df.apply(preprocessing.LabelEncoder().fit_transform)   # This to apply labelencoder to whole dataframe we used apply func.
```

```python
df.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2   # Basically chi2 used for categorical features and target
from sklearn.feature_selection import f_classif    # this is to use annova. Annova is used when featue is continues and target is categorical or vice versa
```

```python
x=df.iloc[:,0:5]   # independent columns
y=df.iloc[:,-1]   # target column .i.e., Drug
```

```python
x
```

```python
bestfeatures = SelectKBest(score_func=f_classif,k=3) # k should be less than or equal to total features
```

```python
fit=bestfeatures.fit(x,y)
dfscores=pd.DataFrame(fit.scores_)
dfscores
```

```python
fit.scores_
```

```python
dfcolumns=pd.DataFrame(x.columns)
dfcolumns
```

```python
featurescores=pd.concat([dfcolumns,dfscores],axis=1)
featurescores
```

```python
featurescores.columns=['Features','Score']
```

```python
featurescores
```

```python
print(featurescores.nlargest(2,'Score'))   # To select top 2 features with best scores
```

# Chi Square used for feature selection

```python
bestfeatures1 = SelectKBest(score_func=chi2,k=5)
```

```python
fit=bestfeatures1.fit(x,y)
dfscores=pd.DataFrame(fit.scores_)
dfscores
```

```python
dfcolumns=pd.DataFrame(x.columns)
featurescores=pd.concat([dfcolumns,dfscores],axis=1)
featurescores.columns=['Features','Score']
featurescores
```

```python
print(featurescores.nlargest(2,'Score')) 
```
