```python
# MAking new dataframe with required features
df1=df[['tenure','age','address','income','ed','employ','equip','churn']]
```

```python
# Type of churn feature is float. we converted it into int type
df1['churn']=df1['churn'].astype('int')
df1.head()
```

```python
# Independent features are seperates in X variable
X=np.asarray(df1[['tenure','age','address','income','ed','employ','equip']])
X[0:5]
```

```python
# Dependent feature/ target feature in Y variable
Y=np.asarray(df1['churn'])
Y[0:5]
```

```python
#Also, We normalize the dataset
```

```python
#We scaling because range of features are different
from sklearn import preprocessing   
X = preprocessing.StandardScaler().fit(X).transform(X)   
X[0:5]    # Here new scale is -2 to 2
```

```python
# Now we split data in Test and Train data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=4)
print('train set: ',x_train.shape,y_train.shape)
print('test set: ',x_test.shape,y_test.shape)
```

```python
# Importing model and fitting with train data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01,solver='liblinear').fit(x_train,y_train)
LR
```

```python
# Predicting Values
yhat=LR.predict(x_test)
yhat
```

```python
# We find Probability of xtest which gives probability getting correct prediction and false prediction
# we can only find this in case of binary class
yhat_prob=LR.predict_proba(x_test)
yhat_prob
```

```python
# Now we find jaccard index for accuracy evaluation , It is applied in logistic because it only worked with probability
from sklearn.metrics import jaccard_score
jaccard_score(y_test,yhat)
```

```python
# Confusion Matrix to check TP , TN , FP , FN Values
cnf_matrix=confusion_matrix(y_test,yhat,labels=[1,0])
cnf_matrix
```

```python
# We plot distribution plot of True Probability and False Probability
import seaborn as sns
sns.distplot(yhat_prob[:,0],hist=False,rug=True)
sns.distplot(yhat_prob[:,1],hist=False,rug=True)
```

```python
# calculate the fpr and tpr for all thresholds of the classification

import sklearn.metrics as metrics

preds=yhat_prob[:,1]
fpr,tpr,threshold=metrics.roc_curve(y_test,preds)
roc_auc=metrics.auc(fpr,tpr)

plt.title('Receiver Operating Characterstic')

plt.plot(fpr,tpr,'b',label='AUC=%0.2f'%roc_auc)
plt.legend(loc='lower right')

plt.plot([0,1],[0,1],'r--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')

plt.show()
```
