```python
# Importing some Modules and Packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Selecting Independent features
x=df[['region', 'tenure', 'age', 'marital', 'address', 'income', 'ed',
       'employ', 'retire', 'gender', 'reside']]
x[0:5]
```

```python
# Selecting Target/Dependent Feature
y=df['custcat']
y[0:5]
```

```python
# Importing train_)test_split function from Model_selection module of sklearn package
# Using the function to get train and test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
print('train set: ',x_train.shape,y_train.shape)
print('test set: ',x_test.shape,y_test.shape)
```

```python
# Importing KNeighborsClassifier from neighbors module
from sklearn.neighbors import KNeighborsClassifier
k=4  # knn Model works on Value of K. So, we give 4 to K

# KNN MOdel is fitted with train data set
neigh=KNeighborsClassifier(n_neighbors=k).fit(x_train,y_train)
neigh
```

```python
#Predictiing values for test features
yhat=neigh.predict(x_test)
yhat[0:5]
```

```python
# Score of KNN model for k=4
neigh.score(x_test,y_test)*100
```

```python
# Importing metric module to use accuarcy_score from sklearn pacakage
from sklearn import metrics
```

```python
# Now we find best value of k for which score of the model is Best

Ks=10
mean_acc=np.zeros((Ks-1))
std_acc=np.zeros((Ks-1))

for n in range(1,Ks):
    neigh=KNeighborsClassifier(n_neighbors=n).fit(x_train,y_train)
    yhat=neigh.predict(x_test)
    mean_acc[n-1]=metrics.accuracy_score(y_test,yhat)
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
    
print(f"Mean Accuracy for k=1 to 9 : {mean_acc}")
print(f"\nStandard deviation of Accuracy for k=1 to 9 : {std_acc}")
```

```python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()   # This Graph is with error
```

```python
# This Graph is without error .i.e., standard deviation
plt.plot(range(1,Ks),mean_acc,'g')
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy is", mean_acc.max(), "with k =", mean_acc.argmax()+1) 
```
